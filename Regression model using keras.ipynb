# Regression  model using keras

## 1.Download and Clean Dataset 

```python
import pandas as pd
import numpy as np
```

### Read the data
**pd.read_csv()** : 

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#to see how many data points are there
concrete_data.shape
```

```python
#check for missing data
concrete_data.describe()
```

### Clean the data

```python
concrete_data.isnull().sum()
```

### Split the data

```python
concrete_data_columns = concrete_data.columns

predictors= concrete_data[concrete_data_columns[concrete_data_columns!='Strength']]
target = concrete_data['Strength']
```

```python
predictors
```

```python
target.head()
```

### Normalize the data

```python
predictors_norm = (predictors - predictors.mean())/predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] #number of predictors
```

## Building the neural network

```python
import keras
```

Since we are using a linear stack of layers , so we would use sequential model


```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def Regression_model():
    
    model = Sequential()
    #add 2 hidden layers and 1 output unit
    model.add(Dense(50,activation='relu',input_shape=(n_cols,)))
    model.add(Dense(50,activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

## Train the neural network

```python
#build the model
model = Regression_model()
```

verbose=2 will just mention the number of epoch like this

```python
#fit the model
model.fit(predictors_norm,target,validation_split=0.3,epochs=100,verbose=2)
```
